````python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def montecarlo2(region, ax, bx, ay, by, N) :
    #limites en x (ax, bx) y en y (ay, by), y el n煤mero de muestra (dardos aventados)
    x = np.random.uniform(ax, bx, N)
    y = np.random.uniform(ay, by, N)
    B = region(x, y) # Bernoulli -> 
    AreA = (bx-ax)*(by-ay) # rea = DELTA de las distancias 
    return np.mean(B)*AreA
````

````python
def bernoulli(x, y) :
    return ((np.exp((-0.05)*(x**2 + y**2)))*(np.sin(x**2 + y**2))) > 0.5
````
